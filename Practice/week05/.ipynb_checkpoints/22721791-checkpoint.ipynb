```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

```

```python
df=pd.read_csv("heart.csv")
df
```

```python
#Danh sách các cột input
columns_eda=df.columns[:-1]
columns_eda
```

```python
# Phân tích phân phối các biến số input
fig=plt.figure(figsize=(16,10))

for i in range(len(columns_eda)):
    colname=columns_eda[i]
    sub=fig.add_subplot(3,5,i+1)
    sns.histplot(data=df, x=colname, kde=True)
```

```python
# Đếm xem số lượng các biến input (nhóm định tính) theo biến mục tiêu
categories_list=['sex', 'cp', 'fbs', 'restecg','exang','slope','ca','thal']
fig=plt.figure(figsize=(20,8))

for i in range(len(categories_list)):
    colname=categories_list[i]
    sub=fig.add_subplot(2,4,i+1)
    sns.countplot(data=df, x=colname, hue="target")
```

```python
# Phân tích biểu đồ Box-plot các biến input (nhóm định lượng) theo biến mục tiêu
numeric_list=['age','trestbps','chol','thalach', 'oldpeak']
fig=plt.figure(figsize=(20,8))

for i in range(len(numeric_list)):
    colname=numeric_list[i]
    sub=fig.add_subplot(2,4,i+1)
    sns.boxplot(data=df, y=colname, x="target")
```

```python
#Chuẩn bị dữ liệu
x=df.iloc[:,:-1].values
y=df[['target']].values
x=x.astype(float)
y=y.astype(float)
```

```python
# Phân chia tập dữ liệu với tỷ lệ 80:20 và random_state=42
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Hiển thị danh sách index của các mẫu trong X_train và X_test
index_train = range(len(X_train))
index_test = range(len(X_test))

print("Danh sách index của các sample trong X_train:", list(index_train))
print("Danh sách index của các sample trong X_test:", list(index_test))
```

```python
# Huấn luyện mô hình
knn=KNeighborsClassifier(n_neighbors=8, algorithm="ball_tree")
knn.fit(X_train, y_train)
```

```python
# e. Đánh giá mô hình trên tập test bằng độ đo accuracy
knn.score(X_test, y_test)
```

```python
# Sử dụng độ đo đánh giá accuracy trên tập train và test để so sánh và chọn lựa K bao 
# nhiêu là tốt nhất cho mô hình K-NN. Sinh viên nhìn biểu đồ để trả lời
neighbors = np.arange(1, 20)
train_accurary = np.empty(len(neighbors))
test_accurary = np.empty(len(neighbors))
#loop over K values
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    #compute training and test data accuracy
    train_accurary[i] = knn.score(X_train, y_train)
    test_accurary[i] = knn.score(X_test, y_test)
    
#generate plot 
plt.plot(neighbors, test_accurary, label = 'Testing dataset Accuracy')
plt.plot(neighbors, train_accurary, label = 'Training dataset Accuracy')

plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
# Tìm giá trị K tốt nhất dựa trên độ chính xác của tập kiểm tra
best_k = neighbors[np.argmax(test_accurary)]
print(f'Giá trị K tốt nhất là: {best_k}')
```

```python
# Huấn luyện lại mô hình với giá trị K tốt nhất
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train, y_train)

# Kiểm tra độ chính xác cuối cùng
final_train_accuracy = best_knn.score(X_train, y_train)
final_test_accuracy = best_knn.score(X_test, y_test)

print(f'Độ chính xác trên tập huấn luyện với K={best_k}: {final_train_accuracy}')
print(f'Độ chính xác trên tập kiểm tra với K={best_k}: {final_test_accuracy}')
```

```python
# In ra danh sách khoảng cách và các hàng xóm từ các mẫu dữ liệu trong tập test từ mô hình 
# ở câu 5
distances, indices = knn.kneighbors(X_test)
print(distances)
print(indices)
```

```python
# 12. Liệt kê danh sách các nhãn dự báo thông qua mô hình của các mẫu dữ liệu input trong tập 
# test. Hãy cho biết kết quả dự báo của mẫu input thứ 5 trong tập test sẽ được dự báo là bao 
# nhiêu?
y_test_prediced = knn.predict(X_test)
y_test_prediced
```

```python
# 14. Hãy đánh giá mô hình trên tập test qua Confussion – Matrix
from sklearn.metrics import confusion_matrix
cfmx = confusion_matrix(y_test, y_test_prediced)
cfmx
```

```python
# 18. Đánh giá mô hình thông qua đồ thị AUC & ROC
from sklearn import metrics
y_pred_proba = knn.predict_proba(X_test)[:,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr, tpr, 'go-', label="AUC="+str(auc))
plt.plot([0,1], [0,1], 'r--')
plt.title("AUC & ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc=4)
plt.fill_between(fpr, tpr, facecolor = 'lightgreen', alpha =0.7)
plt.show()
```
